### Import Library

```python
import pandas as pd
```

### Read Dataset

```python
data_2017 = pd.read_csv("Data/tahun-2017---data-capaian-penanganan-sampah.csv")
data_2018 = pd.read_csv("Data/tahun-2018---data-capaian-penanganan-sampah.csv")
data_2019 = pd.read_csv("Data/tahun-2019---data-capaian-penanganan-sampah.csv")
data_2020 = pd.read_csv("Data/tahun-2020---data-capaian-penanganan-sampah.csv")
data_2021 = pd.read_csv("Data/tahun-2021---data-capaian-penanganan-sampah.csv")
data_gaji_pengeluaran = pd.read_excel("Data/data_gaji_pengeluaran.xlsx")
```

### Menyamakan Nama Column dan Menambahkan Tahun setiap dataframe pada data sampah

```python
data_2017 = data_2017.rename(columns={'Tonase(Ton)': 'Tonase (Ton)'})
data_2017['Tahun'] = 2017
data_2018['Tahun'] = 2018
data_2019['Tahun'] = 2019
data_2020['Tahun'] = 2020
data_2021['Tahun'] = 2021
```

```python
data_2017.head()
```

### Menggabungkan Data Sampah Menjadi 1

```python
data_sampah = pd.concat([data_2017, data_2018, data_2019, data_2020, data_2021], ignore_index=True)
```

```python
data_sampah.head()
```

### Pengecekan Nilai Null

```python
data_sampah.isnull().sum()
```

### Pengecekan Nilai Duplikat

```python
duplikat = data_sampah.duplicated()
baris_duplikat = data_sampah[duplikat]
print(f"Banyak Baris Data Yang Duplikat : {len(baris_duplikat)}")
```

### Export Data Yang Telah Bersih

```python
data_sampah.to_csv('Data/data_sampah.csv', index=False)
```
